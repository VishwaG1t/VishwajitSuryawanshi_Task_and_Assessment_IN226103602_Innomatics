# 1: Social Media – Post Engagement Analyzer

```python

def engagement(likes):
    
    # Calculating total likes
    Total_likes = 0
    
    for l in likes :
        Total_likes += l
        
    print("Total_likes :" , Total_likes)

    # Deciding whether Total likes comes under Viral post or normal
    
    if Total_likes >= 1000 :
        print('Viral Post')
    else:
        print("Normal Engagement")


# likes at different duration on post during the day

time_period = int(input('enter different time intervals in a day for likes :'))

Likes_list = []

for i in range(time_period):
    Like = int(input("enter likes for the time interval :")) ;
    Likes_list.append(Like)

print("Likes at diffrent time intervals : " , Likes_list)
    
engagement(Likes_list)
```

# 3: Agriculture – Rainfall Adequacy Checker

```python

Required_rainfall_level = 172

def rainfall_analyze(rainfall_data):
    
   
# Calculating toatl and avg rainfall
    
    total_rainfall = 0
    for r in rainfall_data :
        total_rainfall += r

    print("total_rainfall :" , total_rainfall)

    Avg_rainfall = total_rainfall / len(rainfall_data)
    print('Avg_rainfall :' , Avg_rainfall)


# Logic to check for adequate or inadequate rainfall
    
    if  Avg_rainfall >= Required_rainfall_level:
        print('Adequate Rainfall')

    else:
        print('Inadequate Rainfall')
    

# Collecting rainfall data

Rainfall = int(input("Rainfall data for given years :"))
rainfall_data = []

for i in range(Rainfall):
    rain = int(input("Enter the amount of rainfall (cm) :"))
    rainfall_data.append(rain)

print(f'Rainfall data for 4 years : {rainfall_data}')

# Calling function to show final analysis
rainfall_analyze(rainfall_data)


```

# 6: Agriculture – Premium Crop Price Filter


```python

def premium_crops(crops):

    Premium_crop = []
    
    for c in crops :
        if c > 2000:
            Premium_crop.append(c)   
            
        else:
            print("regular crop :" , c)

    print("Premium Crops :" , Premium_crop)

crops = [1000,2500,3200,500]
premium_crops(crops)
            




```

```python
def premium_crops(crops):
    premium_crop = [
        c if c > 2000 else print(f"regular : {c}")
        for c in crops
    ]
    
    # Remove None values (because print() returns None)
    premium_crop = [c for c in premium_crop if c is not None]
    
    print("Premium crops:", premium_crop)

crops = [1000, 2500, 3200, 500]
premium_crops(crops)
```
