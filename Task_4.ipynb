# 1: Mobile Recharge Validation System

```python
def valid_recharge():
    while True:
        recharge = int(input('Enter your plan: '))
        
        if recharge < 50:
            print("Minimum recharge is 50")
            continue
        
        if recharge in [199, 299, 399, 599]:
            print("Recharge successful")
            break
        else:
            print("Choose valid recharge plan of 199,299,399,599")
            continue

valid_recharge()
```

# 2: Inventory Reorder Alert System


```python
product = {'chips':100 , 'Chocolate':30 , 'pen':13}

def dukan():
    for i , stock in product.items() :
        if stock < 15 :                   # Checks the Stock quantity
            print(f'{i} : Reorder Alert')
        else:
            print(f'{i} : Stock OK')

dukan()
```

# 3: Student Result Processing System


```python
def Marks(student_marks):

    total_marks = 0
    for marks in student_marks :
        total_marks += marks


    # Calculating average marks student scored
    Avg_marks = total_marks / len(student_marks)
    print("Avg_marks" , Avg_marks)
    
    # Determining whether student passed or failed
    if Avg_marks >= 50 :
       return 'pass'
    else:
        return 'fail'
   
        
student_marks = [10,24,25,40]
Result = Marks(student_marks)
print(Result)


    
```

# 5: Employee Attendance Eligibility 

```python
def attendence(Pappu_Attendence):

    # Logic to count no. of P's
    Count_P = []
    
    for i in Pappu_Attendence :
        if i == 'P' :
            Count_P.append(i)
            
    Attend = len(Count_P)

    # Calculating Attendence Percentage
    Attendence_Percentage = (Attend / len(Pappu_Attendence)) * 100 

    # Determing whether student is eligible or not 
    if Attendence_Percentage >= 75:
        
        return  Attendence_Percentage , 'eligible'
    else:
        return  Attendence_Percentage , 'not eligible'

            


Pappu_Attendence = ['P', 'A', 'A', 'P', 'P']
Student = attendence(Pappu_Attendence)
print(Student)
```

# 6: Password Strength Checker

```python
def check_password_strength():
    while True:
        password = input("Enter password: ")

        # Rule 1: Minimum length
        if len(password) < 8:
            print("Weak Password: Must be at least 8 characters long.\n")
            continue

        has_digit = False
        has_special = False

        # Loop through each character
        for char in password:
            if char.isdigit():
                has_digit = True
            if char in "@#$":
                has_special = True

        # Check conditions
        if not has_digit:
            print("Weak Password: Must contain at least one digit.\n")
            continue

        if not has_special:
            print("Weak Password: Must contain at least one special character (@, #, $).\n")
            continue

        # If all conditions pass
        print("Strong Password ✅")
        break


check_password_strength()
```
