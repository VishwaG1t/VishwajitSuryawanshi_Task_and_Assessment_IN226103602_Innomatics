# Task 1: Replace Hardcoded Prompts

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

def explain_topic_template():
    return PromptTemplate.from_template("Explain {topic} in simple terms for beginners")
prompt1 = explain_topic_template()
print(prompt1.format(topic="Quantum Computing"))
```

# Task 3: Prompt Variations Engine

```python

multi_template = PromptTemplate.from_template("Explain {topic} for {audience} in a {tone} tone")
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]
for case in test_cases:
    print(multi_template.format(**case))
print("\n")
```

# Task 3: Prompt Variations Engine

```python

templates = {
    "Teaching": PromptTemplate.from_template("Explain {topic} clearly step by step"),
    "Interview": PromptTemplate.from_template("Ask 3 questions about {topic}"),
    "Storytelling": PromptTemplate.from_template("Explain {topic} as a story")
}
for name, template in templates.items():
    print(f"{name}: {template.format(topic='Machine Learning')}")
print("\n")
```

# Task 4: Chat Prompt Template System

```python

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert {role}."),
    ("user", "My topic is {topic}.")
])

roles = ["teacher", "interviewer", "motivator"]

for role in roles:
    messages = chat_template.format_messages(role=role, topic="Neural Networks")
    print(f"Role: {role.capitalize()}")

    for msg in messages:
        print(f"  {msg.type.capitalize()}: {msg.content}")
print("\n")
```

#  Task 5: Input Validation Layer

```python

def validate_inputs(audience: str, tone: str):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "casual"

    if audience not in valid_audiences or tone not in valid_tones:
        print(f"  [Warning] Invalid inputs. Defaulting to audience='{final_audience}', tone='{final_tone}'.")

    return final_audience, final_tone

print("  Testing valid: ", validate_inputs("expert", "formal"))
print("  Testing invalid:", validate_inputs("aliens", "angry"))
print("\n")
```

# Task 6: Prompt Generator App

```python

def generate_prompt(topic: str, audience: str, tone: str, style: str):
    val_audience, val_tone = validate_inputs(audience, tone)
    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} tone, clearly step by step.",
        "interview": "Conduct an interview about {topic} for a {audience} audience, keeping a {tone} tone.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }
    selected_template_str = style_templates.get(style.lower(), style_templates["teaching"])
    prompt = PromptTemplate.from_template(selected_template_str)
    return prompt.format(topic=topic, audience=val_audience, tone=val_tone)

print(generate_prompt(topic="Neural Networks", audience="beginner", tone="fun", style="storytelling"))
print("\n")
```
