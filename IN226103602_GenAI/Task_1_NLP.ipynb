**Q: What is the difference between "Love" and "love" in NLP?**

Ans: Same word, different case—if text isn’t lowercased, NLP may treat them as separate tokens.

**Q: What happens if stopwords are not removed?**

Ans: Model may get noisy data and focus on less meaningful words, reducing performance.

**Q: Provide two real-world scenarios where removing stopwords can be harmful.**

Ans: When removing stopwords is harmful :-

1. Sentiment analysis: Removing “not” flips meaning (“not good” → “good”).

2. Search queries/questions: Removing words like “how”, “where” can change intent and reduce accuracy.

**Q:** **Explain the difference between stemming and lemmatization.**

Ans: : Stemming chops words to crude roots; lemmatization returns proper dictionary forms.

```python
import re
from collections import Counter
```

```python


# ==========================================
# Task 2 & Task 7: Preprocessing Function
# ==========================================


def preprocess_text(text):
    # Task 7: Error Handling (Empty strings, non-strings)
    if not isinstance(text, str) or not text.strip():
        return [], ""

    # Task 2: Lowercase the text
    text = text.lower()

    # Task 2: Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www.\S+|\S+@\S+', '', text)

    # Task 2: Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation and emojis (keeping only letters and whitespace)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Task 2: Handle repeated characters (Reduces 3 or more identical chars to 1)
    # Note: "goooood" becomes "god" with basic regex. To keep "good", a dictionary mapping is usually required in advanced NLP.
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Task 2: Remove extra spaces and tokenize
    tokens = text.split()

    # Task 2: Remove very short tokens (length <= 2) with exceptions
    keep_words = {"no", "not", "ok"} # Added 'ok' based on sample inputs
    clean_tokens = [word for word in tokens if len(word) > 2 or word in keep_words]

    # Join back into a clean sentence
    clean_sentence = " ".join(clean_tokens)

    # Task 7: If the input was "only emojis" or "only numbers", clean_tokens will safely be an empty list [] here.
    return clean_tokens, clean_sentence

# ==========================================
# Task 6: Build Full Pipeline
# ==========================================
def full_pipeline(text_list):
    results = {
        "tokens": [],
        "clean_sentences": []
    }

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        results["tokens"].append(tokens)
        results["clean_sentences"].append(clean_sentence)

    return results

# ==========================================
# Task 3, 4, & 5: Execution and Analytics
# ==========================================
if __name__ == "__main__":
    sample_inputs = [
        "Get 100% FREE access now!!!",
        "I absolutely looooved this product 😍😍",
        "Worst service ever... 0/10",
        "Call me at 98763210",
        "This is THE best course!!!",
        "Visit https://openai.com now!",
        "Nooooo this is baaad!!!",
        "OK OK OK I got it",
        "Win $$$ now!!! Limited offer!!!",
        "I am not happy with this",
        "",           # Task 7 Test: Empty String
        "😍😍😍",     # Task 7 Test: Only emojis
        "1234 5678"   # Task 7 Test: Only numbers
    ]

    # Run the pipeline
    pipeline_results = full_pipeline(sample_inputs)
    all_tokens_combined = []

    print("--- TASK 3 & 4: STRESS TESTING & ANALYTICS ---")

    for i, original in enumerate(sample_inputs):
        tokens = pipeline_results["tokens"][i]
        sentence = pipeline_results["clean_sentences"][i]
        all_tokens_combined.extend(tokens)

        # Task 4 Analytics calculations
        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

        print(f"\nOriginal: '{original}'")
        print(f"Cleaned Tokens: {tokens}")
        print(f"Cleaned Sentence: '{sentence}'")
        print(f"Analytics -> Total: {total_tokens} | Unique: {unique_tokens} | Avg Length: {avg_length:.2f}")

    print("\n--- TASK 5: FREQUENCY ANALYSIS ---")

    token_counts = Counter(all_tokens_combined)

    print("Top 10 most frequent words:")
    for word, count in token_counts.most_common(10):
        print(f" - {word}: {count}")

    print("\nTop 5 least frequent words:")

    # Reverse the most_common list to get the least frequent
    for word, count in token_counts.most_common()[:-6:-1]:
        print(f" - {word}: {count}")
```

```python
# Task 4 Analysis Questions Ans:

# Calculate noise (characters lost) for each sentence
noise_scores = []
for original, cleaned in zip(sample_inputs, pipeline_results["clean_sentences"]):
    chars_lost = len(original) - len(cleaned)
    noise_scores.append((chars_lost, original))

# Find the maximum
max_noise_value, noisiest_sentence = max(noise_scores)
print(f"Most Noise: '{noisiest_sentence}' (Lost {max_noise_value} characters)")


# Count the length of the cleaned token lists
token_counts = []
for original, tokens in zip(sample_inputs, pipeline_results["tokens"]):
    token_counts.append((len(tokens), original))

# Find the maximum
max_token_count, most_meaningful_sentence = max(token_counts)
print(f"Most Meaningful: '{most_meaningful_sentence}' ({max_token_count} tokens retained)")
```
